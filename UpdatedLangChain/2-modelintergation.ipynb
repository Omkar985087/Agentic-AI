Models Integration with openAI,google gemini and groq

```python
import os 
from dotenv import load_dotenv
load_dotenv()

os.environ["OPENAI_API_KEY"]=os.getenv("OPENAI_API_KEY")
os.environ["GROQ_API_KEY"]=os.getenv("GROQ_API_KEY")
os.environ["GOOGLE_API_KEY"]=os.getenv("GOOGLE_API_KEY")
```

```python
from langchain.chat_models import init_chat_model

model=init_chat_model("groq:llama-3.3-70b-versatile")
model
```

```python
#invoke the model
response=model.invoke("Hello how are you")
response
```

google gemini model intergation

```python
model=init_chat_model("google_genai:gemini-2.5-flash-lite")
response=model.invoke("Why do parrots talk")
response.content
```

Streaming

```python
model.invoke("write me a 200 words paragraph on artificial intelligence")
```

```python
for chunk in model.stream("write me a 200 words paragraph on artificial intelligence"):
    print(chunk.text,end="|",flush=True)
```

```python
for chunk in model.stream("why do parrots have colorfull feathers?"):
    print(chunk.text,end="|",flush=True)
```

Batch

```python
responses=model.batch([
    "why do parrots have colorfull feathers",
    "how do airplanes fly",
])

for response in responses:
    print(response)
```
